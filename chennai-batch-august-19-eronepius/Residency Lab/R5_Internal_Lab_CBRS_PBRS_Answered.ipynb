# Popularity Based Recommendation System

### About Dataset

Anonymous Ratings on jokes.

1. Ratings are real values ranging from -10.00 to +10.00 (the value "99" corresponds to "null" = "not rated").

2. One row per user

3. The first column gives the number of jokes rated by that user. The next 100 columns give the ratings for jokes 01 - 100.

# Q10 Read the dataset(jokes.csv)

Take care about the header in read_csv() as there are no column names given in the dataset. 

```python
jokes = pd.read_csv('./jokes.csv')
jokes.head()
```

# Q11 Consider `ratings` named dataframe with only first 200 rows and all columns from 1(first column is 0) of dataset

```python
ratings = jokes.iloc[0:200, 1:]
```

```python
ratings.shape
```

```python
ratings.head()
```

# Q12 Change the column indices from 0 to 99

```python
ratings.rename(columns={x:y for x,y in zip(ratings.columns,range(0,len(ratings.columns)))}, inplace=True)
```

```python
ratings.head()
```

# Q13 In the dataset, the null ratings are given as 99.00, so replace all 99.00s with 0
Hint: You can use `ratings.replace(<the given value>, <new value you wanted to change with>)`

```python
ratings.replace(99.00,0, inplace=True)
```

# Q14 Normalize the ratings and save them in `ratings_diff` variable

```python
# importing standardscalar to normalize the values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
print(scaler.fit(ratings))
print(scaler.mean_)
rating_diff = scaler.transform(ratings)
```

### Popularity based recommendation system

# Q15  Find the mean for each column  in `ratings_diff` i.e, for each joke
Consider all the mean ratings and find the jokes with highest mean value and display the top 10 joke IDs.

```python
def findmeanrating(rating_diff):
    meanrating = []
    for item in rating_diff:
        meanrating.append(item.mean())
    return(meanrating)
```

```python
meanrating = findmeanrating(rating_diff)
```

```python
ratings['meanrating'] = meanrating
```

```python
ratings.head()
```

```python
highRated = ratings[ratings['meanrating'] > 0.6]
```

```python
highRated.sort_values(['meanrating'],ascending=False, inplace=True)
```

```python
highRated.head()
```

```python
highratedJokesid = highRated.index
```

```python
print("Top to Jokes is are : \n")
print(highratedJokesid.tolist()[:11])
```
